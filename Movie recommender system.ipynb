```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
#Merge above two datasets
movies = movies.merge(credits,on='title')
movies.head()
```

```python
#Remove unessential columns
#columns that will  help me in creating tags for content based recommendation system.(look through public point of view, not yours)
#genre, id, keywords, overview, cast, crew
movies = movies[['id','title','genres','keywords','overview','cast','crew']]
movies.head()
```

```python
#Now our goal is to create a dataframe with columns
#Movie_id
#title
#tags(?)
#check for missing data
movies.isnull().sum()
```

```python
#drop rows with missing values
movies.dropna(inplace=True)
```

```python
#Look for duplicate rows
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ['Action','Adventure','Fantasy','Science Fiction']
import ast
def convert(obj):
    name = []
    for i in ast.literal_eval(obj):
        name.append(i['name'])
    return name
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(5)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(5)
```

```python
def convert3(obj):
    name = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            name.append(i['name'])
            counter+=1
        else:
            break
    return name
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(5)
```

```python
def director(obj):
    name = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            name.append(i['name'])
            break
    return name
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(5)
```

```python
 movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(5)
```

```python
#Remove space transformation
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
movies.head()
```

```python
data = movies[['id','title','tags']]
data.head()
```

```python
data['tags'] = data['tags'].apply(lambda x:" ".join(x))
```

```python
data.head()
```

```python
data['tags'] = data['tags'].apply(lambda x:x.lower())
```

```python
data.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemm(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stemm)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = data[data['title'] == movie].index[0]
    movies_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])
    for i in movies_list:
        print(data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
